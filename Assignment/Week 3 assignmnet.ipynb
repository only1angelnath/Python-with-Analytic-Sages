Build a Python script that fetches, processes, and displays the top 100 cryptocurrency market data using the CoinGecko API.

```python
import requests
import os
import pandas as pd
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("Coingecko_API") 

# 1. Use the 'markets' endpoint for the Top 100
url = "https://api.coingecko.com/api/v3/coins/markets"

# 2. Set vs_currency and per_page=100
params = {
    "vs_currency": "usd",
    "include_tokens": "top",
    "order": "market_cap_desc",
    "per_page": 100,
    "page": 1,
    "sparkline": False,
    "precision": 4
}

headers = {
    "accept": "application/json",
    "x-cg-demo-api-key": api_key 
}

try: 
    response = requests.get(url, params=params, headers=headers)
    response.raise_for_status()
    
    data = response.json()
    
    if data:
        df = pd.DataFrame(data)
        
        print(f"✅ Success! Fetched {len(df)} tokens.")
        print(df[['name', 'symbol', 'current_price', 'market_cap', 'last_updated']].head()) 
    else:
        print("❌ No data returned.")
    
except Exception as e:
    print(f"❌ Error: {e}")
```

```python
data = response.json()

# 2. Create a DataFrame
df = pd.DataFrame(data)

# 3. Clean up the data
# Convert 'last_updated' directly (it's already a timestamp string)
df['updated_at'] = pd.to_datetime(df['last_updated'])

# Define the columns (Added the missing comma and the supply columns you wanted)
columns_to_keep = [ 'market_cap_rank', 'name', 'symbol', 'current_price', 'market_cap', 'total_volume', 'updated_at']

# Filter the DataFrame
df_filtered = df[columns_to_keep]

# 4. Save to CSV
file_name = "top_100_token_by_mcap.csv"
df_filtered.to_csv(file_name, index=False)

print(f"✅ Success! File saved as: {file_name}")
print(f"Total rows saved: {len(df_filtered)} rows")

# 5. Show the result
print(df_filtered.head())
```
