```python
from web3 import Web3
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv()
INFURA = os.getenv("INFURA_URL")

print(f'infura imported as {INFURA}')

w3 = Web3(Web3.HTTPProvider(INFURA))
print(f'Is connected to infura?: {w3.is_connected()}')
```

### 1. Pick up an address and use get balance meethod, use dot get balance method and add the address to get it. Get balance address

```python
#checksumming the address
def turn_to_checksum(address: str) -> str:
    if not Web3.is_checksum_address(address):
        return Web3.to_checksum_address(address)
    return address

WALLET_ADDRESS_RAW = "0x09f96D8da0Ac189a1eaD263E1ebEB25c76fCB0Bc"
WALLET_ADDRESS = turn_to_checksum(WALLET_ADDRESS_RAW)
print(f'Is the address valid?: {Web3.is_checksum_address(WALLET_ADDRESS)}') #this checks if the address is valid
```

```python
balance = w3.eth.get_balance(WALLET_ADDRESS)
print(f'Balance in Wei: {balance}')
balance_in_eth = w3.from_wei(balance, 'ether')
print(f'Balance in Ether: {balance_in_eth} ETH')
```

### 2. Filter transaction by address

```python
wallet = WALLET_ADDRESS.lower()
latest_block = w3.eth.block_number

txs = []

# scan last 100 blocks (adjust as needed)
for i in range(latest_block - 100, latest_block + 1):
    block = w3.eth.get_block(i, full_transactions=True)
    
    for t in block.transactions:
        from_addr = (t.get('from') or '').lower()
        to_addr = (t.get('to') or '').lower()
        
        if wallet in (from_addr, to_addr):
            txs.append(t)

print(f"Found {len(txs)} transactions for {WALLET_ADDRESS} in the last 100 blocks with the latest block being {latest_block}.")
```

### 3. Get latest block and details using the latest method.

```python
from datetime import datetime

block = w3.eth.get_block('latest', full_transactions=True)

print("=" * 40)
print(f"Block Number: {block.number}")
print(f"Time: {datetime.fromtimestamp(block.timestamp)}")
print(f"Miner: {block.miner}")
print(f"Gas Used: {block.gasUsed}/{block.gasLimit}")
print(f"Transactions: {len(block.transactions)}")
print("=" * 40)

for tx in block.transactions[:5]:
    print("\nTransaction:")
    print(f"Hash: {tx['hash'].hex()}")
    print(f"From: {tx['from']}")
    print(f"To: {tx['to']}")
    print(f"Value: {w3.from_wei(tx['value'], 'ether')} ETH")
```
